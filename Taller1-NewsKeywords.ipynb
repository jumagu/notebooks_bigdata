Primero importamos la librerías y módulos de python necesarias

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd
from pathlib import Path  
import glob
```

Creamos una variable para establecer la ruta de la cual obtendremos los archivos .txt

Luego con el módulo glob creamos una lista con todas las rutas de cada archivo .txt

```python
directory_path = "C:/Trabajos/Octavo/BigData/datosTallerReto/tech"
text_files = glob.glob(f'{directory_path}/*.txt')
text_files
```

Ahora con Path().stem sacamos el último elemto de la ruta, en este caso el nombre del archivo

```python
text_titles = [Path(text).stem for text in text_files]
```

```python
text_titles
```

Aquí primero inicializamos el TfidfVectorizer, luego se ejecuta usando nuestro vector de noticias.

```python
tfidf_vectorizer = TfidfVectorizer(input='filename', stop_words='english')
```

```python
tfidf_vector = tfidf_vectorizer.fit_transform(text_files)
```

Creamos un dataframe para guardar y mostrar nuestros datos.

```python
tfidf_df = pd.DataFrame(tfidf_vector.toarray(), index=text_titles, columns=tfidf_vectorizer.get_feature_names_out())
```

```python
tfidf_slice = tfidf_df[['program', 'engine', 'navigate', 'technology', 'industry']]
tfidf_slice.sort_index().round(decimals=2)
```

Aquí reorganizamos el dataframe para que las palabras clave estén en filas en lugar de columnas.

```python
tfidf_df.stack().reset_index()
```

```python
tfidf_df = tfidf_df.stack().reset_index()
```

```python
tfidf_df = tfidf_df.rename(columns={0:'tfidf', 'level_0': 'document','level_1': 'term', 'level_2': 'term'})
```

La organización de las palabras clave va ser de acuerdo al tf-idf más alto de cada documento, se organizarán de forma ascendente y se tomarán las primeras 5 palabras con más frecuencia.

```python
tfidf_df.sort_values(by=['document','tfidf'], ascending=[True,False]).groupby(['document']).head(5)
```

Finalmente hay que hacer que se muestren las palabras clave de cada documento, para esto usaremos la **biblioteca de visualización de datos altair**

Básicamente en esta parte se configura el aspecto visual, el orden y la forma en que se mostrarán los datos en la tabla de Altair

```python
import altair as alt
import numpy as np
```

```python
top_tfidf = tfidf_df.sort_values(by=['document','tfidf'], ascending=[True,False]).groupby(['document']).head(5)
top_tfidf_plusRand = top_tfidf.copy()
top_tfidf_plusRand['tfidf'] = top_tfidf_plusRand['tfidf'] + np.random.rand(top_tfidf.shape[0])*0.0001
```

```python
base = alt.Chart(top_tfidf_plusRand).encode(
    x = 'rank:O',
    y = 'document:N'
).transform_window(
    rank = "rank()",
    sort = [alt.SortField("tfidf", order="descending")],
    groupby = ["document"],
)
```

```python
heatmap = base.mark_rect().encode(color = 'tfidf:Q')
```

```python
text = base.mark_text(baseline='middle').encode(text = 'term:N', color = alt.condition(alt.datum.tfidf >= 0.23, alt.value('black'), alt.value('black')))
```

Y este es el resultado final de de las palabras clave de las noticas usando la fórma del tf-idf.

```python
(heatmap + text).properties(width = 400)
```
